# Cheminformatic reactions - definitions, usage, and applications
`polymerist` provides a number of utilities and novel algorithms for handling chemical reactions between functional groups, which allows one to approach chemical rearrangements, especially as pertains to polymerizations, ina "context-free" manner; a single reaction template can be applied to an entire chemical class of molecules. these methods are useful for numberous applications, including:
* Enumeration of polymer repeat units
* Automatic detection of monoer-to-mechanism compatibility (synthsis pathway identification)
* Auomatic detection of autopolymerization reactions
* Scalable, high-throughput reaction-based workflows (scales as number of chemical classes, NOT as number of molecules)

Here we provide several tutorials on the reaction-related funcitonality `polymerist` supplies, including:
* 0.1 - Setting up reaction templates from bond cycles
* 0.2 - Tracing atoms and bonds through a reaction
* 0.3 - forming and fragmenting reaction adducts
* 0.4 - Generating and storing chemical fragments 
* 0.5 - Enumeration of compatible reactants via DISCERNMENT
* 0.6 - [Performing reactions in series to enumerate all possible chemical fragments](0.6-modified_mpd_tmc_fragmentation.ipynb)
