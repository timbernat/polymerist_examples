# Miscellaneous
Here, we provide demos of additional functionality `polymerist` provides which don't neatly fit into any of   
the prior MD pipeline demos or which simply aren't as polished as some of the other tutorials provided here
* [Detecting ring piercing events](./ring-piercing.ipynb) - demonstration of a novel algorithm introduced in `polymerist` for automated, general-purpose ring piercing detection during polymer conformer generation, formally called PINPRICS (Planar Intersection Nomography for Pierced Ring Identification via Cut Sets)
* [JSONification](./jsonification.ipynb) - overview of the API and convenience classes polymerist provides to let you write basically any object to JSON
* [Boron with the OpenFF toolkit](./openff_boron_demo.py) - quick dmeo of how to load boron-containing molecules into the OpenFF toolkit (not supported by Sage FF at time of writing, needs some fanagling as shown here)
